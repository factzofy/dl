```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([apsit])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
for sentence in apsit.split('/n'):
  print(sentence)
```

```python
for sentence in apsit.split('\n'):
  print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
input_sequences = []
for sentence in apsit.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
[len(x) for x in input_sequences]
```

```python
max([len(x) for x in input_sequences])
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
#zero padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
padded_input_sequences[:,:-1]
```

```python
X = padded_input_sequences[:,:-1]
```

```python
padded_input_sequences[:,-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=555)
```

```python
y.shape
```

```python
y[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(555, 100, input_length=50))
model.add(LSTM(150))
model.add(Dense(555, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
# predict next 15 words
import time
text = "omkar"

for i in range(15):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=50, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
